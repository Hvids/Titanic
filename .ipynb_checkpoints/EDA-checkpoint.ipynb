```python
def decribe_unique_count(all_df, train_df, test_df):
    for column in all_df.columns:
        print('__________________________________________')
        print(f'Column = {column}' )
        try:
            unique_all_count = np.count_nonzero(all_df[column].unique())
            unique_train_count = np.count_nonzero(all_df[column].unique())
            unique_test_count = np.count_nonzero(all_df[column].unique())
            print(f'all: \n count_unique = { unique_all_count } | size = {all_df.shape[0]}')
            print(f'train: \n count_unique =  { unique_train_count } | size = {train_df.shape[0]}')
            print(f'test: \n count_unique =  { unique_test_count } | size = {test_df.shape[0]}')
        except:
            print('Error')
            continue
decribe_unique_count(all_df, train_df, test_df)
```

## Раьотаем с новыми признаками


```python
def make_name_appeal(name):
    if not name.find('Master.') == -1:
        return 'Master.'
    elif not name.find('Mr.') == -1:
        return 'Mr.'
    elif not name.find('Mrs.') == -1:
        return 'Mrs.'
    elif not name.find('Miss.') == -1:
        return 'Miss.'
    elif not name.find('Ms.') == -1:
        return 'Ms.'
    elif not name.find('Don.') == -1:
        return 'Don.'
    elif not name.find('Rev.') == -1:
        return 'Rev.'
    elif not name.find('Dr.') == -1:
        return 'Dr.'
    elif not name.find('Major.') == -1:
        return 'Major.'
    elif not name.find('Mme.') == -1:
        return 'Mme.'
    elif not name.find('Lady.') == -1:
        return 'Lady.'
    elif not name.find('Sir.') == -1:
        return 'Sir.'
    elif not name.find('Mlle.') == -1:
        return 'Mlle.'
    elif not name.find('Col.') == -1:
        return 'Col.'
    elif not name.find('Capt.') == -1:
        return 'Capt.'
    elif not name.find('Dona.') == -1:
        return 'Dona.'
    elif not name.find('Countess.') == -1:
        return 'Countess.'
    elif not name.find('Jonkheer.') == -1:
        return 'Jonkheer.'
    else:
        return 'Nan'

    
train_df['Appeal'] = train_df.Name.apply(make_name_appeal)
test_df['Appeal'] = test_df.Name.apply(make_name_appeal)
```

```python
def make_cabin_floor(cabin):
    if pd.isna(cabin):
        return np.nan
        
    else:
        cabin_list = cabin.split(' ')
        cabin_ = cabin_list[0]
        cabin_str = cabin_[0]
        return cabin_str
        
def make_cabin_count_namber(cabin):
    if pd.isna(cabin):
        return np.nan
    else:
        cabin_list = cabin.split(' ')
        cabin_ = cabin_list[0]
        cabin_int = cabin_[1:]
        return len(cabin_int)

train_df['Floor'] = train_df.Cabin.apply(make_cabin_floor)
test_df['Floor'] = test_df.Cabin.apply(make_cabin_floor)
    
train_df['Count_number'] = train_df.Cabin.apply(make_cabin_count_namber)
test_df['Count_number'] = test_df.Cabin.apply(make_cabin_count_namber)
    
    
```

```python
def make_ticket_char(ticket):
    ticket_list = ticket.split(' ')
    if len(ticket_list) == 1:
        return np.nan
    else:
        return ticket_list[0]

train_df['Ticket_char'] = train_df.Ticket.apply(make_ticket_char)
test_df['Ticket_char'] = test_df.Ticket.apply(make_ticket_char)
```

```python
def make_family(row):
    parch = row.Parch
    sibsp = row.SibSp
    return parch + sibsp

train_df['Family'] = train_df.apply(make_family,axis=1)
test_df['Family'] = test_df.apply(make_family,axis=1)
    
    
```

# Создаем новые датасеты для разных моделей

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from copy import deepcopy

def make_dataset_tree(data, numerical, categorical):
    data = deepcopy(data)
    ss = StandardScaler()
    data[numerical] = ss.fit_transform(data[numerical])
    return data
def make_dataset_liner(data, numerical, categorical):
    data = deepcopy(data)
    ss = StandardScaler()
    one = OneHotEncoder()
    hot_data = one.fit_transform(data[categorical]).toarray()
    hot_data = pd.DataFrame(hot_data, columns=one.get_feature_names(categorical))
    data[numerical] = ss.fit_transform(data[numerical])
    data.drop(categorical,axis = 1)
    data = pd.concat([data, hot_data], axis = 1)
    return data
```

```python
categorical_features = ['Pclass', 'Sex', 'Embarked', 'Appeal', 'Floor', 'Count_number', 'Ticket_char']
numerical_features = ['Age', 'SibSp', 'Parch', 'Fare', 'Family']
```

```python
all_df_knn_liner = make_dataset_liner(all_df_knn, numerical_features, categorical_features)
all_df_knn_tree =  make_dataset_tree(all_df_knn, numerical_features, categorical_features)
```

```python
all_df_imp_liner = make_dataset_liner(all_df_imp, numerical_features, categorical_features)
all_df_imp_tree = make_dataset_tree(all_df_imp, numerical_features, categorical_features)
```

```python
def make_train_test(data, size_train, target):
    data = deepcopy(data)
    train = data.iloc[:size_train]
    train['Survived'] = target
    test = data.iloc[size_train:]
    return train, test
```

```python
size_train = train_df.shape[0]
train_knn_liner, test_knn_liner = make_train_test(all_df_knn_liner,size_train, train_df.Survived)
train_knn_tree, test_knn_tree = make_train_test(all_df_knn_tree,size_train, train_df.Survived)
```

```python
train_imp_liner, test_imp_liner = make_train_test(all_df_imp_liner,size_train, train_df.Survived)
train_imp_tree, test_imp_tree = make_train_test(all_df_imp_tree,size_train, train_df.Survived)
```
